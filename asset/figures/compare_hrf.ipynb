```python
# Import Libraries
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import nibabel as nib
import pandas as pd
```

```python
# Init var 
analysis_folder = '/project/4180000.19/multirat_stim/scratch/rabies_test/'
first_level_path = analysis_folder+'first_level/'
second_level_path = analysis_folder+'second_level/'

# --- Load data --- 

 # - Rat HRF - 
gamma_hrf_first = mpimg.imread(first_level_path+'rat_hrf_gamma_diff/image/stat_map_sub-0200100_ses-1.png')
gamma_hrf_second = mpimg.imread(second_level_path+'rat_hrf_gamma_diff/image/stat_map_dataset-02001_z_score.png')
gamma_hrf_model = pd.read_csv(first_level_path+'rat_hrf_gamma_diff/model/model_sub-0200100_ses-1.csv')

dd_hrf_first = mpimg.imread(first_level_path+'rat_hrf_delay_disp/image/stat_map_sub-0200100_ses-1.png')
dd_hrf_second = mpimg.imread(second_level_path+'rat_hrf_delay_disp/image/stat_map_dataset-02001_z_score.png')
dd_hrf_model = pd.read_csv(first_level_path+'rat_hrf_delay_disp/model/model_sub-0200109_ses-1.csv')

 # - Default HRF - 
glover_hrf_first = mpimg.imread(first_level_path+'glover/image/stat_map_sub-0200100_ses-1.png')
glover_hrf_second = mpimg.imread(second_level_path+'glover/image/stat_map_dataset-02001_z_score.png')
glover_hrf_model = pd.read_csv(first_level_path+'glover/matrix/matrix_sub-0200100_ses-1.csv')
glover_hrf_model = glover_hrf_model.iloc[:, 0]

spm_hrf_first = mpimg.imread(first_level_path+'spm/image/stat_map_sub-0200100_ses-1.png')
spm_hrf_second = mpimg.imread(second_level_path+'spm/image/stat_map_dataset-02001_z_score.png')
spm_hrf_model = pd.read_csv(first_level_path+'spm/matrix/matrix_sub-0200100_ses-1.csv')
spm_hrf_model = spm_hrf_model.iloc[:, 0]
```

```python
#--- Compare Default HRF --- 

fig = plt.figure(figsize=(20, 10))
gs = fig.add_gridspec(nrows=3, ncols=2, hspace=0.5, wspace=0.01, height_ratios=[2, 4, 4], width_ratios=[1, 1])  #hspace (height space) and wspace (width space)

# Create Three Axes Objects
ax1 = fig.add_subplot(gs[0, 0])
ax1.plot(glover_hrf_model)
ax1.set_ylabel('HRF amplitude')
ax1.set_xlabel('Time (s)')
ax1.set_title('Default glover HRF Model')
ax1.set_axis_off()

ax2 = fig.add_subplot(gs[0, 1])
ax2.plot(spm_hrf_model)
ax2.set_xlabel('Time (s)')
ax2.set_title('Default spm HRF Model')
ax2.set_axis_off()

ax3 = fig.add_subplot(gs[1, 0])
ax3.imshow(glover_hrf_first)
ax3.set_title('First Level Analysis')
ax3.set_axis_off()

ax4 = fig.add_subplot(gs[1, 1])
ax4.imshow(spm_hrf_first)
ax4.set_title('First Level Analysis')
ax4.set_axis_off()

ax5 = fig.add_subplot(gs[2, 0])
ax5.imshow(glover_hrf_second)
ax5.set_title('Second Level Analysis')
ax5.set_axis_off()

ax6 = fig.add_subplot(gs[2, 1])
ax6.imshow(spm_hrf_second)
ax6.set_title('Second Level Analysis')
ax6.set_axis_off()


plt.savefig('/project/4180000.19/multirat_stim/scratch/plots/compare_default_hrf.svg', format='svg', dpi=1200) 
```

```python
# -- Compare Rat HRF -- 

fig = plt.figure(figsize=(20, 10))

gs = fig.add_gridspec(nrows=3, ncols=2, hspace=0.5, wspace=0.01, height_ratios=[2, 4, 4], width_ratios=[1, 1])  #hspace (height space) and wspace (width space)

# Create Three Axes Objects
ax1 = fig.add_subplot(gs[0, 0])
ax1.plot(gamma_hrf_model)
ax1.set_ylabel('HRF amplitude')
ax1.set_xlabel('Time (s)')
ax1.set_title('2 Gammas Function Rat HRF Model')
ax1.set_axis_off()

ax2 = fig.add_subplot(gs[0, 1])
ax2.plot(dd_hrf_model)
ax2.set_xlabel('Time (s)')
ax2.set_title('2 Parameters Rat HRF Model')
ax2.set_axis_off()

ax3 = fig.add_subplot(gs[1, 0])
ax3.imshow(gamma_hrf_first)
ax3.set_title('First Level Analysis')
ax3.set_axis_off()

ax4 = fig.add_subplot(gs[1, 1])
ax4.imshow(dd_hrf_first)
ax4.set_title('First Level Analysis')
ax4.set_axis_off()

ax5 = fig.add_subplot(gs[2, 0])
ax5.imshow(gamma_hrf_second)
ax5.set_title('Second Level Analysis')
ax5.set_axis_off()

ax6 = fig.add_subplot(gs[2, 1])
ax6.imshow(dd_hrf_second)
ax6.set_title('Second Level Analysis')
ax6.set_axis_off()

plt.savefig('/project/4180000.19/multirat_stim/scratch/plots/compare_rat_hrf.svg', format='svg', dpi=1000) 

```
