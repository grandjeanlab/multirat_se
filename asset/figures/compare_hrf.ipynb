```python
# Import Libraries
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import nibabel as nib
import pandas as pd

#https://towardsdatascience.com/create-panel-figure-layouts-in-matplotlib-with-gridspec-7ec79c218df0
```

```python
# --- Rat HRF ---

# Init var 
rat_analysis_folder = '/project/4180000.19/multirat_stim/scratch/rabies_test/first_level_custom_rat/'

rat_hrf_first = mpimg.imread(rat_analysis_folder+'rat_hrf/image/stat_map_sub-0200100_ses-1.png')
rat_hrf_second = mpimg.imread(rat_analysis_folder+'rat_hrf/sec_level_map/stat_map_dataset-02001_z_score.png')
rat_hrf_model = pd.read_csv(rat_analysis_folder+'rat_hrf/model/hrf_model_sub-0200109_ses-1.csv')

generic_hrf_first = mpimg.imread(rat_analysis_folder+'rat_hrf_generic/image/stat_map_sub-0200100_ses-1.png')
generic_hrf_second = mpimg.imread(rat_analysis_folder+'rat_hrf_generic/sec_level_map/stat_map_dataset-02001_z_score.png')
generic_hrf_model = pd.read_csv(rat_analysis_folder+'rat_hrf_generic/model/generic_hrf_model_sub-0200109_ses-1.csv')


#Â --- Default HRF ---
default_analysis_folder = '/project/4180000.19/multirat_stim/scratch/rabies_test/abstract_ISMRM/'

glover_hrf_first = mpimg.imread(default_analysis_folder+'glover/stat_map_sub-0200100_ses-1.png')
glover_hrf_second = mpimg.imread(default_analysis_folder+'glover/2nd_level/stat_map_dataset-02001_z_score.png')
glover_hrf_model = pd.read_csv(default_analysis_folder+'glover/matrix/matrix_sub-0200100_ses-1.csv')
glover_hrf_model = glover_hrf_model.iloc[:, 0]

spm_hrf_first = mpimg.imread(default_analysis_folder+'spm/stat_map_sub-0200100_ses-1.png')
spm_hrf_second = mpimg.imread(default_analysis_folder+'spm/2nd_level/stat_map_dataset-02001_z_score.png')
spm_hrf_model = pd.read_csv(default_analysis_folder+'spm/matrix/matrix_sub-0200100_ses-1.csv')
spm_hrf_model = spm_hrf_model.iloc[:, 0]
```

```python

fig = plt.figure(figsize=(20, 10))

gs = fig.add_gridspec(nrows=3, ncols=2, hspace=0.5, wspace=0.2, height_ratios=[2, 4, 4], width_ratios=[1, 1])  #hspace (height space) and wspace (width space)


# Create Three Axes Objects
ax1 = fig.add_subplot(gs[0, 0])
ax1.plot(glover_hrf_model)
ax1.set_ylabel('HRF amplitude')
ax1.set_xlabel('Time (s)')
ax1.set_title('Default glover HRF Model')
#ax1.set_axis_off()

ax2 = fig.add_subplot(gs[0, 1])
ax2.plot(spm_hrf_model)
ax2.set_xlabel('Time (s)')
ax2.set_title('Default spm HRF Model')
#ax2.set_axis_off()

ax3 = fig.add_subplot(gs[1, 0])
ax3.imshow(glover_hrf_first)
ax3.set_title('First Level Analysis')
ax3.set_axis_off()

ax4 = fig.add_subplot(gs[1, 1])
ax4.imshow(spm_hrf_first)
ax4.set_title('First Level Analysis')
ax4.set_axis_off()

ax5 = fig.add_subplot(gs[2, 0])
ax5.imshow(glover_hrf_second)
ax5.set_title('Second Level Analysis')
ax5.set_axis_off()

ax6 = fig.add_subplot(gs[2, 1])
ax6.imshow(spm_hrf_second)
ax6.set_title('Second Level Analysis')
ax6.set_axis_off()
```
